```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
#from plotly.offline import plot


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('results/square_optimization_grades.csv')
```

```python
data = data.drop('Unnamed: 0', axis = 1)
data
```

```python
data["alternative_id"] = data["alternative_id"].astype('category')
data = data[data['alpha']== 0.5]
data = data.sort_values(by=['fun_val'] , ascending=False)
data
```

```python
fig = px.bar(data, x='alternative_id', y='fun_val')
fig.show()
```

```python
ord_list = list(data['alternative_id'])
```

```python
fig = px.bar(data, x='alternative_id', y='fun_val',
             hover_data=['optimal_grade'],
              height=400, width=1500)
fig.update_layout( xaxis={'categoryorder':'array', 'categoryarray': ord_list })
fig.show()

```

```python
data.sort_values(by = 'fun_val', ascending = False)
```

```python
fig = px.bar(data, y='fun_val', x='alternative_id', text='optimal_grade')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```
