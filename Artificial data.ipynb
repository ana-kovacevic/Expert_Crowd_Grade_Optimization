```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
# low low low

n_cases = 50
n_range = 11
def create_df(df_list):
    
    dfs = []
    for exp, cro, case in df_list:
        exp_pd = pd.DataFrame({'votes':exp})
        exp_pd['group'] = 'expert'
        exp_pd['case'] = case
        exp_pd['id'] = np.arange(len(exp))
        dfs.append(exp_pd)
        
        cro_pd = pd.DataFrame({'votes':cro})
        cro_pd['group'] = 'crowd'
        cro_pd['case'] = case
        cro_pd['id'] = np.arange(len(cro))
        dfs.append(cro_pd)
        
        
    total_df = pd.concat(dfs)
    
    return(total_df)




exp_1 = np.array([1,2,1])
cro_1 = np.random.choice(np.arange(1, n_range), n_cases, p=[0.7, 0.2, 0.05, 0, 0.05, 0, 0, 0, 0, 0])


exp_2 = np.array([1,2,1])
cro_2 = np.random.choice(np.arange(1, n_range), n_cases, p=[0, 0, 0, 0, 0, 0, 0.05, 0.1, 0.15, 0.7])


exp_4 = np.array([9,10,10])
cro_4 = np.random.choice(np.arange(1, n_range), n_cases, p=[0.2, 0, 0.2, 0.2, 0, 0, 0.1, 0, 0, 0.3])


exp_6 = np.array([1,5,7])
cro_6 = np.random.choice(np.arange(1, n_range), n_cases, p=[0, 0, 0, 0, 0, 0.05, 0, 0.1, 0.15, 0.7])

exp_7 = np.array([1,5,9])
cro_7 = np.random.choice(np.arange(1, n_range), n_cases, p=[0.25, 0, 0.25, 0, 0.25/2, 0.25, 0, 0, 0.25/2, 0])

exp_8 = np.array([1,5,10])
cro_8 = np.random.choice(np.arange(1, n_range), n_cases, p=[0.2/3, 0, 0.4, 0, 0.2/3, 0, 0, 0, 0.4, 0.2/3])






df_list = [(exp_1, cro_1, 'case_1'), (exp_2, cro_2, 'case_2'), (exp_4, cro_4, 'case_4'),\
           (exp_6, cro_6, 'case_6'), (exp_7, cro_7, 'case_7'), (exp_8, cro_8, 'case_8')]


total_df = create_df(df_list)
plt.close()
#plt.figure(figsize = (14,4))
#sns.boxplot(x="case", y="votes", hue = 'group', data=total_df, palette="Set3", linewidth=2.5)
#sns.catplot(x="group", y="votes", col = 'case', data=total_df, kind = 'swarm', palette="Set3",  linewidth=2.5, height = 4., aspect = .7)
sns.catplot(x="group", y="votes", col = 'case', data=total_df, palette="Set3", kind = 'box', linewidth=2.5, height = 4., aspect = .7)
sns.catplot(x="group", y="votes", col = 'case', data=total_df, palette="Set3", kind = 'swarm', linewidth=2.5, height = 4., aspect = .7)


plt.show()
```

```python
total_df.to_csv('results/artifical_data_example.csv')
```
